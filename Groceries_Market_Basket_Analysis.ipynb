```python
# Import the necessary package
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# Import the dataset
df = pd.read_csv('../Data/Groceries_dataset.csv')
df.head(10)
```

```python
df.shape
```

```python
# Combine Member_number with Date to create purchase id as unique identifier or Receipt
df['purchase_id'] = df['Member_number'].astype(str) + '_' + df['Date'].astype(str)
df.head()
```

```python
df['purchase_id'].nunique()
```

```python
df2 = pd.crosstab(df['purchase_id'], df['itemDescription'])
df2.head()
```

```python
# Check if there's any value that is bigger than 1
(df2 > 1).any().any()
```

```python
# Transform any single item's purchase qty to 1 if it is more than 1
basket_input = df2.where(df2 == 0, 1)
basket_input.head()
```

```python
# Check if there's any value that is bigger than 1
(basket_input > 1).any().any()
```

```python
frequent_itemsets = apriori(basket_input, min_support=0.001, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift")
rules.head()
```

```python
rules.sort_values(["support", "confidence","lift"],axis = 0, ascending = False).head(8)
```
